# 2. Build a regression model to predict the profit from the given data

Here, all the correlation values are so close to 1. Hence, there is no need for feature reduction.

```python
data.columns
```

```python
y=data['Profit(in Thousands)']
X=data.drop(['Profit(in Thousands)','Date'],axis=1)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
```

```python
lr=linear_model.LinearRegression()
```

```python
model=lr.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
print(y_test)
```

```python
print(predictions)
```

The obtained values are the "profit generated values" by giving 25% of dataset model and comparing with the y_test data, 
we can see that almost all the values in both cases are same.

```python
from sklearn.metrics import mean_squared_error
```

```python
print("MSE is:",mean_squared_error(y_test,predictions))
```

The Mean Square Error is very small as it takes value 4.416*10^-11. Hence, the regression line perfectly fits and accurate.

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test, predictions)
```

```python
r2
```

The R square value is exactly equal to 1. From here also, we can conclude that the regression line perfectly fits.

# 3. Predict the profit generated by the company,if they wish to spend the following amount for advertisement

```python
make_pred=[[290,312,309,298,320]]
model.predict(make_pred).item()
```
